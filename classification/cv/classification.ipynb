```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define reader 
def load_mnist(path, prefix='train'):
    import os
    import gzip
    import numpy as np

    """Load MNIST data from `path`"""
    labels_path = os.path.join(path,
                               '%s-labels-idx1-ubyte.gz'
                               % prefix)
    images_path = os.path.join(path,
                               '%s-images-idx3-ubyte.gz'
                               % prefix)

    with gzip.open(labels_path, 'rb') as lbpath:
        labels = np.frombuffer(lbpath.read(), dtype=np.uint8,
                               offset=8)

    with gzip.open(images_path, 'rb') as imgpath:
        images = np.frombuffer(imgpath.read(), dtype=np.uint8,
                               offset=16).reshape(len(labels), 784)

    return images, labels
```

```python
# 读取本地训练数据, 并查看数据的 shape
data_path='./datasets/fashion'
train_images, train_labels = load_mnist(data_path, prefix='train')
test_images, test_labels = load_mnist(data_path, prefix='t10k')
print("train_images shape: {}, train_labels shape: {}".format(train_images.shape, train_labels.shape))
print("test_images shape: {}, test_labels shape: {}".format(test_images.shape, test_labels.shape))
```

```python
# 对标签 0-9 的数字做映射
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# reshape，每张图片都是 28 * 28 的像素组成
train_images = train_images.reshape(60000, 28, 28)
test_images = test_images.reshape(10000, 28, 28)
```

```python
# 展示其中一个图片
plt.figure(figsize=(5,3))
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
# 归一化: 将这些值缩小至 0 到 1 之间，然后将其馈送到神经网络模型。为此，请将这些值除以 255。请务必以相同的方式对训练集和测试集进行预处理：

train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
plt.figure(figsize=(10,10), dpi=74)
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

```python
# 构建模型

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10)
])
```

```python
# 编译模型
# loss 评估预测和真实的差距，反向求导可以计算每个参数的梯度信息
# optimizer 根据梯度，更新参数，目的是降低 loss 
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
# 训练模型
model.fit(train_images, train_labels, epochs=10, verbose=1)
```

```python
# 评估准确率
# 观察模型在测试集的上表现
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('Test accuracy:', test_acc)
```

```python
# 进行预测
probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])
```

```python
predictions = probability_model.predict(test_images)
```

```python
# 通过绘制图表的方式，来直观感受一些模型预测的结果

# 打印图片以及与对应真实标签和预测标签
def plot_image(i, predictions, test_labels, test_imges):
    prediction, test_label, test_image = predictions[i], test_labels[i], test_images[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(test_image, cmap=plt.cm.binary)
    predict_index = np.argmax(prediction)
    predict_probability = np.max(prediction) * 100
    color = "red" if predict_index == test_label else "blue"
    xlabel_str = "{} {:2f}% ({})".format(class_names[predict_index], 
                                         predict_probability,
                                         class_names[test_label])
    plt.xlabel(xlabel_str, color=color)
    

# 通过柱状图来查看预测效果
def plot_histogram(i, predictions, test_labels):
    prediction, test_label = predictions[i], test_labels[i]
    plt.grid(False)
    plt.xticks(range(10))
    plt.yticks([])
    # 设置 y 轴的显示范围
    plt.ylim([0, 1])
    histogram_plot = plt.bar(x=range(10), height=prediction, color='gray')
    histogram_plot[np.argmax(prediction)].set_color("red")
    histogram_plot[test_label].set_color("blue")
    
```

```python
# 选取一张测试集的图片
i = 28
# 设置画布
plt.figure(figsize=(8, 3), dpi=74)
# 图 1
plt.subplot(1, 2, 1)
plot_image(i, predictions, test_labels, test_images)
# 图二
plt.subplot(1, 2, 2)
plot_histogram(i, predictions, test_labels)
plt.show()
```
